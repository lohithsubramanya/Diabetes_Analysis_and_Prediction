#### Analysis is not required for the outlier information as it is practically impossible to have 0 for blood hg and mass
#### I will get the count of such outliers.


* Also low BP leads to less oxygen and blood to the parts of body that includes kidneys, brain and heart. 
* Low BP directly does not kill a person, it gradually attacks each vital organ which leads to a bigger damage.
* That means a low BP condition can exist but it is a very critical condition.

```python
df.loc[df['plas'] < 1,'plas'] = np.nan
df.loc[df['skin'] < 1,'skin'] = np.nan
df.loc[df['mass'] < 1,'mass'] = np.nan
df.loc[df['blood hg'] < 25,'blood hg'] = np.nan
df.loc[df['serum mu'] < 1,'serum mu'] = np.nan
```

# Information from a doctor with respect to serum insulin: 

* Fasting:  < 25 mIU/L
* 30 minutes after glucose administration: 30-230 mIU/L
* 1 hour after glucose administration: 18-276 mIU/L
* 2 hour after glucose administration: 16-166 mIU/L

#### Therefore, it is impossible to have 0 serum insulin levels

```python
#### Graphs after adding null values


nrows, ncols = 4, 2

fig = plt.figure(figsize=(15,15))    
for i,col in zip(range(1,9), bio_cols):
    ax = fig.add_subplot(nrows, ncols, i)
    ax.set_title(col)
    df[col].hist()
plt.subplots_adjust(hspace = 0.5)
plt.show()
```

# Data Imputing

```python
from sklearn.linear_model import LinearRegression
full_vals = ['preg','pedi','age']
missing_col = ['plas','blood hg','skin','serum mu', 'mass']
for col in missing_col:
    not_null_index = df.ix[df[col].notnull(),col].index
    X = df.ix[not_null_index][full_vals].values
    Y = df.ix[not_null_index][col].values
    clf = LinearRegression(fit_intercept=True)
    clf.fit(X,Y)
    #clf.predict(df.ix[df[col].isnull()][full_vals].values)
    if len(not_null_index) > 0:
        df.ix[df[col].isnull(), col] = clf.predict(df.ix[df[col].isnull()][full_vals].values)
```

```python
### Logistic Regression

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
names = df.columns[:-1]
array = df.values
X = array[:,0:8]
Y = array[:,8]
test_size = 0.20
seed = 7
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size,
random_state=seed)
model = LogisticRegression()
model.fit(X_train, Y_train)
result = model.score(X_test, Y_test)
print(result)
```

```python
from sklearn.metrics import classification_report
matrix = classification_report(model.predict(X_test), Y_test)
```
